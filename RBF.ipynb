```python
#### CREATE train_RBF.pkl and test_RBF.pkl from train_PCA.pkl###

"""STUDENT CODE"""
import pickle
import numpy as np
import math


train_cluster=pickle.load(open( "Kmean_cluster.pkl", "rb" ),encoding ='latin1')
train_data=pickle.load(open( "train_PCA.pkl", "rb" ),encoding ='latin1')
test_data=pickle.load(open( "test_PCA.pkl", "rb" ),encoding ='latin1')



#please do not re-generate the file 
#from sklearn.cluster import KMeans
#generated by KMeans(n_clusters=1000,init='random', random_state=10).fit(train_data['data'])

```

```python
train_feature = train_data['data']
train_target = train_data['target']
test_feature = test_data['data']
test_target = test_data['target']
```

```python
cluster_means = np.zeros([train_feature.shape[1],1000], dtype=float)
cluster_sigmas = []
```

```python
cluster_dict = {}
for i in range(len(train_cluster)):
    if train_cluster[i] in cluster_dict:
        cluster_dict[train_cluster[i]].append(train_feature[i])
    else:
        temp = [train_feature[i]]
        cluster_dict[train_cluster[i]] = temp
```

```python
for i in range(1000):
    cluster_means[:,i] = np.mean(cluster_dict[i],axis=0)
```

```python
for i in range(np.shape(cluster_means)[1]):
    distance = 1000000
    for x in train_feature:
        temp = np.sum(np.square(cluster_means[:,i]-x))
        if temp > 0 and temp < distance:
            distance = temp
    cluster_sigmas.append(distance)
```

```python
train_RBF_data = []
for x in train_feature:
    x_RBF = []
    for i in range(1000):
        temp = x - cluster_means[:,i]
        temp = -np.dot(temp,temp)*1.0/(2*cluster_sigmas[i])
        temp = math.exp(temp)
        x_RBF.append(temp)
    train_RBF_data.append(x_RBF)
    
```

```python
test_RBF_data = []
for x in test_feature:
    x_RBF = []
    for i in range(1000):
        temp = x - cluster_means[:,i]
        temp = -np.dot(temp,temp)*1.0/(2*cluster_sigmas[i])
        temp = math.exp(temp)
        x_RBF.append(temp)
    test_RBF_data.append(x_RBF)
```

```python
test_RBF=dict()
test_RBF['data']=np.array(test_RBF_data)
test_RBF['target']=test_data['target']
```

```python
train_RBF=dict()
train_RBF['data']=np.array(train_RBF_data)
train_RBF['target']=train_data['target']
```

```python
# f = open("train_RBF.pkl", "wb")
# pickle.dump(train_RBF, f)
# f.close()
pickle.dump(train_RBF, open("train_RBF.pkl", "wb"))
pickle.dump(test_RBF, open("test_RBF.pkl", "wb"))
```
